```python
import os
import cv2
import numpy as np
from PIL import Image

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')#学習データ

img_dir_path = input('画像フォルダのパスを入力してください：')#

#img_dir_path = '/Users/user/Desktop/drive-download'
pictures = os.listdir(img_dir_path)#フォルダリストの中身をリスト化

if '.DS_Store' in pictures:
    pictures.remove('.DS_Store')

if 'Thumbs.db' in pictures:
    pictures.remove('Thumbs.db')
    
print(pictures)

os.mkdir(os.path.join(img_dir_path,'output'))

i = 0

while i < len(pictures):

    img = cv2.imread(os.path.join(img_dir_path,pictures[i]))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.2, minSize = (250,250), maxSize = (1600,1600))


#    for (x, y, w, h) in faces:
#        cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255))
#        resized_img = cv2.resize(img,(450, 300))
#        cv2.imshow("test.jpg", resized_img)
#        cv2.waitKey(0)
#        cv2.destroyAllWindows 
    print(pictures[i])
    print(faces)

#顔認証に失敗した場合
    if len(faces) == 0:
        img = cv2.imread(os.path.join(img_dir_path,pictures[i]))


# 回転角の指定
        for j in range(-15,15):#-４５°〜４５°まで画像を回転
            angle = 5*j
            scale = 1.0

            height, width = img.shape[:2]
#回転の中心を指定                          
            center = (int(width/2), int(height/2))
        
            trans = cv2.getRotationMatrix2D(center, angle , scale)
#アフィン変換
            img = cv2.warpAffine(img, trans, (width,height))
        
        
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.2, minSize = (250,250), maxSize = (1600,1600))
            if len(faces) > 0:
                print('回転させた結果認識!' + str(faces))
                
                
                
                break
            
                
            
        
        
        
        
        #im_crop = img.crop((100, 200, 300, 400))
        #im_crop.save(os.path.join(img_dir_path,'output',pictures[i])+"【fail】.jpg", quality=100) #dpiも設定してください dpi=(300, 300))
        cv2.imwrite(os.path.join(img_dir_path,'output',pictures[i])+"【fail】.jpg", img)
    
#顔認証に成功した場合
    elif len(faces) > 0:
        face_max = faces[faces[:, 2].argmax()]

        print(face_max)

        img = Image.open(os.path.join(img_dir_path,pictures[i]))
        width, height = img.size

        print(face_max[0])

#クロップ位置の定義（顔を中心に正方形に切り取る）
        X1 = face_max[0]+0.5*face_max[2]-0.5*height
        Y1 = (face_max[1]+0.5*face_max[3]-0.5*width)*0.5#
        X2 = X1 + height
        Y2 = Y1 + width


#横写真だった場合の条件
        if X1 < 0:
            X1 = 0
            X2 = X1 + height
        
        if X2 > width:
            X2 = width
            X1 = X2 - height

        if Y1 < 0:
            Y1 = 0
            Y2 = Y1 + width

        if Y2 > height:
            Y2 = height
            Y1 = Y2 - width

        if width > height:#横写真の場合、以下のようにクロップ
            #im_crop = img.crop((X1, 0, X2, height))
            im_crop = img.crop((X1 + height*17.5/615, 0, X2 - height*17.5/615, height))
            im_resize = im_crop.resize((580,615))
            im_resize.save(os.path.join(img_dir_path,'output',pictures[i]), quality=100, dpi=(300, 300)) #dpiも設定してください dpi=(300, 300))
            
        elif width < height:#縦写真だった場合、
            im_crop = img.crop((0, Y1, width, Y2 + width*35/615))
            im_resize = im_crop.resize((580,615))
            im_resize.save(os.path.join(img_dir_path,'output',pictures[i]), quality=100, dpi=(300, 300)) #dpiも設定してください dpi=(300, 300))

    i = i + 1 

print('END')

```
