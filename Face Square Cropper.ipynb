```python
import os
import cv2
from PIL import Image

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')#学習データ

#img_dir_path = input('画像フォルダのパスを入力してください：')#

img_dir_path = '/Users/user/Desktop/drive-download'
pictures = os.listdir(img_dir_path)#フォルダリストの中身をリスト化
#pictures.remove('.DS_Store')

print(pictures)
print(pictures[0])

#os.listdir(os.path.join(img_dir_path,pictures[i]))#
#img = cv2.imread(pictures[0])
img = cv2.imread(os.path.join(img_dir_path,pictures[0]))
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face_cascade.detectMultiScale(gray, scaleFactor = 1.2, minSize = (300,300), maxSize = (1200,1200))

"""
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 0, 255))
    resized_img = cv2.resize(img,(450, 300))
    cv2.imshow("test.jpg", resized_img)
    cv2.waitKey(0)
    cv2.destroyAllWindows 
"""    
    
print(faces)

#顔認証に失敗し場合
if len(faces) == 0:
    img = Image.open(os.path.join(img_dir_path,pictures[0]))
    im_crop = img.crop((100, 200, 300, 400))
    im_crop.save('crop_tekito.jpg', quality=95)

    
#顔認証に成功した場合
elif len(faces) > 0:
    face_max = faces[faces[:, 2].argmax()]

    print(face_max)

    img = Image.open(os.path.join(img_dir_path,pictures[0]))
    width, height = img.size

    print(face_max[0])

#クロップ位置の定義（顔を中心に正方形に切り取る）
    X1 = face_max[0]+0.5*face_max[2]-0.5*height
    Y1 = (face_max[1]+0.5*face_max[3]-0.5*width)*0.5#
    X2 = X1 + height
    Y2 = Y1 + width

    print()
    print(X1)
    print(Y1)
    print(X2)
    print(Y2)

#横写真だった場合の条件
    if X1 < 0:
        X1 = 0
        X2 = X1 + height
        
    if X2 > width:
        X2 = width
        X1 = X2 - height

    if Y1 < 0:
        Y1 = 0
        Y2 = Y1 + width

    if Y2 > height:
        Y2 = height
        Y1 = Y2 - width

    if width > height:#横写真の場合、以下のようにクロップ
        im_crop = img.crop((X1, 0, X2, height))
        im_crop.save('crop1.jpg', quality=95)

    elif width < height:#縦写真だった場合、
        im_crop = img.crop((0, Y1, width, Y2))
        im_crop.save('crop2.jpg', quality=95)



```
